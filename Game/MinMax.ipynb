## Min-Max Algorithm :
- Maximizer level at depth = 0
- Minimizer level at depth = 0

```python
#Maximizer

import math

def minimax(state, depth, maximizing_player, node_index):
    print(f"{'Maximizer' if maximizing_player else 'Minimizer'} at depth {depth}, node {node_index}")

    if is_terminal(state) or depth == 0:
        utility_value = utility(state, node_index)
        print(f"Returning utility {utility_value} from node {node_index} at depth {depth}")
        return utility_value

    if maximizing_player:
        max_eval = -math.inf
        for i, action in enumerate(actions(state)):
            eval = minimax(result(state, action), depth - 1, False, node_index * 2 + i)
            max_eval = max(max_eval, eval)
            print(f"Maximizer updating max_eval={max_eval} at depth {depth}, node {node_index}")
        return max_eval
    else:
        min_eval = math.inf
        for i, action in enumerate(actions(state)):
            eval = minimax(result(state, action), depth - 1, True, node_index * 2 + i)
            min_eval = min(min_eval, eval)
            print(f"Minimizer updating min_eval={min_eval} at depth {depth}, node {node_index}")
        return min_eval

def utility(state, node_index):
    return state[node_index]

def is_terminal(state):
    return False

def actions(state):
    return [0, 1]

def result(state, action):
    return state

max_depth = int(input("Enter the maximum depth of the tree: "))
num_leaves = 2 ** max_depth

print(f"Enter {num_leaves} leaf node values:")
values = [int(input()) for _ in range(num_leaves)]

optimal_value = minimax(values, max_depth, True, 0)

print("\nThe optimal value is:", optimal_value)
```
